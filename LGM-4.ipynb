```python
#importing libraries
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
image = cv2.imread( "C:\\Users\\Hii Ayush\\Pictures\\Camera Roll\\flower.png" )
```

Loading the image

```python
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

 Converting the Original Image to GrayScale

```python
gray_image = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
invert_image = cv2.bitwise_not(gray_image)
plt.figure(figsize=(8,8))
plt.imshow(invert_image, cmap="gray")
plt.axis("off")
plt.title("Imverted Image")
plt.show()
```

Blurring the Image to make the sketch smooth

```python
smoothing_image = cv2.GaussianBlur(invert_image ,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(smoothing_image, cmap="gray")
plt.axis("off")
plt.title("Smoothing Image")
plt.show()
```

Converting the image to Pencil Sketch

```python
sketch_image = cv2.divide(gray_image ,255 - smoothing_image, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(sketch_image, cmap="gray")
plt.axis("off")
plt.title("Sketch Image")
plt.show()
```

Conversion in one frame

```python
plt.figure(figsize=(20,15))
plt.subplot(1,5,1)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(sketch_image, cmap="gray")
plt.axis("off")
plt.title("Sketch Image")
plt.show()
```
